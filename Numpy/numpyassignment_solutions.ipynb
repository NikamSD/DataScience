```python
data_type = [('x', 'i4'), ('y', 'i4')]
points9_2 = np.array([(1,2), (3,4), (5,6), (2,5)], dtype=data_type)
points9_2
```

```python
points_len = len(points9_2)
points_len
```

```python
distances = np.zeros((points_len, points_len))
for row in range(points_len):
    for column in range(points_len):
        if row != column:

            dx = points9_2['x'][column] - points9_2['x'][row]
            print(f"dx: {dx}\n")
            dy = points9_2['y'][column] - points9_2['y'][row]
            print(f"dy: {dy}\n")
            distances[row, column] = np.sqrt(dx**2 + dy**2)
print(f"Euclidean Distance between points: {distances}")
```

```python
## Available solution for 9.2

# Create a structured array with fields 'x' and 'y'
data_type = [('x', 'i4'), ('y', 'i4')]
data = np.array([(1, 2), (3, 4), (5, 6)], dtype=data_type)
print("Original array:")
print(data)

# Compute the Euclidean distance between each pair of points
distances = np.sqrt((data['x'][:, np.newaxis] - data['x'])**2 + (data['y'][:, np.newaxis] - data['y'])**2)
print("Euclidean distances:")
print(distances)
```

```python
arr10_2 = np.random.randint(1, 21, size=(3,3))
arr10_2
```

```python
##mask the diagonal elements
masked_diag = ma.masked_array(arr10_2, mask=np.eye(3, dtype=bool))
masked_diag
```

```python
unmasked_mean=masked_diag.mean()
print(unmasked_mean)
```

```python
#replace maked elements with mean of unmasked elements
new_masked_array = masked_diag.filled(unmasked_mean)
new_masked_array
```
