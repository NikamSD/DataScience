```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
sns.scatterplot(x='year', y='passengers', data=flights)
plt.title('Year vs Passengers')
plt.show()
```

```python
sns.lineplot(x='year', y='passengers', data=flights)
plt.title('Year vs Passengers') 
plt.show()
```

```python
# Categorical Plot
# BAr plot
sns.barplot(x='year', y='passengers', data=flights)
plt.title('Year vs Passengers') 
plt.show()
```

```python
##Box plot
sns.boxplot(x='month', y='passengers', data=flights)
plt.title('Month vs Passengers')
plt.show()
```

```python
## Violin Plot
sns.violinplot(x='month', y='passengers', data=flights)
```

```python
## Histogram
sns.histplot(flights['passengers'], bins=10)
plt.title('Passengers Histogram')
plt.show()
```

```python
## Histogram
sns.histplot(flights['passengers'], bins=20)
plt.title('Passengers Histogram')
plt.show()
```

```python
##KDE plot

sns.kdeplot(flights['passengers'], fill=True)
plt.title('Passengers KDE Plot')
plt.show()
```

```python
## Pairplot
sns.pairplot(flights)
```

```python
healthexp = sns.load_dataset('healthexp')
healthexp
```

```python
correlation = healthexp[['Year', 'Spending_USD', 'Life_Expectancy']].corr()
correlation
```

```python
#Heatmap
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import pandas as pd

sales_data = pd.read_csv('sales_data.csv')
five_data = sales_data.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Product Name', y='Total Revenue', data=five_data, estimator=sum)
plt.title('Total Revenue by Product Name')
plt.xlabel('Product Name')
plt.ylabel('Total Revenue')
#plt.xticks(rotation=45)
plt.show()
```

```python
## Plot total sales by Region
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y="Total Revenue",data=sales_data,estimator=sum)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```
