```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

```python
x = np.random.randint(1, 10, size=(5)) 
x =sorted(x.tolist())
print(x)

y = [i**2 for i in x]
print(y)
```

```python
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

```python
## Create a customized line plot
plt.plot(x,y, color='orange', linestyle='--',marker='o', markerfacecolor='blue', markersize=10,linewidth=2)
plt.grid(True)
```

```python
x = np.random.randint(1, 10, size=(5)) 
x =sorted(x.tolist())
print(x)

y1 = [i**2 for i in x]
y2= [i**3 for i in x]
print(y1)
print(y2)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(2,2,1)
plt.plot(x, y1, color='green', marker='o')
plt.title('X vs Y1 (X squared)')
plt.xlabel('X-axis')
plt.ylabel('Y1-axis')   
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(x, y2, color='red', marker='s')
plt.title('X vs Y2 (X cubed)')
plt.xlabel('X-axis')        
plt.ylabel('Y2-axis')   
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(y1, x, color='blue', marker='o')
plt.title('Y1 vs X (X squared)')
plt.xlabel('Y1-axis')        
plt.ylabel('X-axis')   
plt.grid(True)  

plt.subplot(2,2,4)
plt.plot(y2, x, color='purple', marker='s')
plt.title('Y2 vs X (X cubed)')      
plt.xlabel('Y2-axis')        
plt.ylabel('X-axis')   
plt.grid(True)

plt.show()
```

```python
## Bar plot

categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 10, size=5)
values = values.tolist()
print(values)
```

```python
plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot of Categories vs Values')
plt.show()
```

```python
## Histogram -> Frequency distribution of data
data = np.random.randint(1, 10, size = 20)
data = data.tolist()
print(data)
```

```python
plt.hist(data, bins=range(1,10), color='lightgreen', edgecolor='black')
```

```python
##create a scatter plot
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
plt.scatter(x, y, color='red', marker='x')
plt.xlabel('X-axis')        
plt.ylabel('Y-axis')   
plt.title('Simple Scatter Plot')
plt.grid(True)
plt.show()
```

```python
##pie chart
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [20, 35, 30, 15]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0,0.1,0,0)  # explode 2nd slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, shadow=True, autopct='%1.1f%%')
plt.title('Pie Chart of Categories')
plt.show()
```

```python
## Sales Data Visualization
sales_data_df =pd.read_csv('sales_data.csv')
sales_data_df.head()
```

```python
sales_data_df.info()
```

```python
total_sales_revenue = sales_data_df.groupby('Product Category')['Total Revenue'].sum() 
total_sales_revenue
```

```python
#bar plot
total_sales_revenue.plot(kind='bar', color='orange', edgecolor='black')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.title('Total Sales Revenue by Product Category')
plt.show()
```

```python
sales_trend = sales_data_df.groupby('Date')['Total Revenue'].sum()
sales_trend
```

```python
sales_trend = sales_data_df.groupby('Date')['Total Revenue'].sum().reset_index()
sales_trend
```

```python
plt.plot(sales_trend['Date'], sales_trend['Total Revenue'], color='purple')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.title('Sales Trend Over Time')
plt.grid(True)
plt.show()
```
