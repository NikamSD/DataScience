##### dropna in detail

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [np.nan, 2, 3, 4],
    'C': ['x', None, 'z', 'w']
})
df
```

```python
# Drop rows containing any NA

df_clean = df.dropna()
df_clean
```

```python
# Drop rows only when every column is NA
df.dropna(how='all')
```

```python
# Drop columns that contain any NA
df.dropna(axis=1)
```

```python
# Only look at columns A and B when deciding to drop rows
df.dropna(subset=['A', 'B'])
# Rows with NA in C remain if A and B are OK
```

```python
# Keep rows that have at least 2 non-null values
df.dropna(thresh=2)
```

```python
# Modify df directly
df.dropna(inplace=True)
# Now df is changed. Do NOT write df = df.dropna() if you used inplace=True
df
```

```python
s = pd.Series([1, None, 2, np.nan, 3])
s_clean = s.dropna()  # drops None and np.nan, returns Series [1,2,3]
s_clean
```

```python
# When using boolean mask keep alignment semantics in mind:
mask = df['A'] > 1
df.loc[mask]            # selects rows by label using mask
print(mask)
# df.iloc[mask] would raise or behave incorrectly  convert to values if using iloc
df.iloc[mask.values]   # positional selection using boolean array
```

```python
df.dropna(axis=1, thresh=3)
df
```

```python
# Prefer assignment rather than chained indexing or inplace on views
df = df.dropna(subset=['A','B'])  # reassign result to df (clear and explicit)
df
```

##### fillna in detail

This section shows common and advanced `fillna()` patterns: scalar fills, column-wise fills (using a dict), forward/backward fills (`ffill`/`bfill`), using `thresh`-style logic with `fillna` + aggregation (mean/median), `inplace` vs assignment, and `combine_first` for filling from another DataFrame.

```python
import pandas as pd
import numpy as np

df_fill = pd.DataFrame({
    'A': [1, np.nan, 3, np.nan],
    'B': [np.nan, 'b', None, 'd'],
    'C': [10, 20, np.nan, 40]
})

# show original
print('Original:')
print(df_fill)
```

```python
# 1) Scalar fill: fill all NaNs with a single value
df_scalar = df_fill.fillna(0)
print('Scalar fill (0):')
print(df_scalar)
```

```python
# 2) Column-wise fill using a dict (different fill for each column)
df_dict = df_fill.fillna({'A': 99, 'B': 'missing', 'C': df_fill['C'].mean()})
print('Column-wise fill (dict):')
print(df_dict)
```

```python
# 3) Forward-fill and backward-fill (propagate non-NA values)

df_fill = pd.DataFrame({
    'A': [1, np.nan, 3, np.nan],
    'B': [np.nan, 'b', None, 'd'],
    'C': [10, 20, np.nan, 40]
})

# show original
print('Original:')
print(df_fill)

df_ffill = df_fill.fillna(method='ffill')  # forward fill
df_bfill = df_fill.fillna(method='bfill')  # backward fill
print('Forward fill:')
print(df_ffill)
print('Backward fill:')
print(df_bfill)

# limit parameter: only propagate up to `limit` consecutive NaNs
df_ffill_limit = df_fill.fillna(method='ffill', limit=1)
print('Forward fill with limit=1:')
print(df_ffill_limit)
```

```python
# 4) Fill using aggregation (mean/median) per-column
# Use assignment to keep results; df_fill.fillna(df_fill.mean()) fills numeric columns with column means
df_mean = df_fill.fillna(df_fill.mean(numeric_only=True))
print('Fill numeric columns with column mean:')
print(df_mean)

# For a single column:
df_single = df_fill.copy()
df_single['A'] = df_single['A'].fillna(df_single['A'].mean())
print('Fill column A with its mean:')
print(df_single)
```

```python
# 5) Inplace vs assignment
df_inplace = df_fill.copy()
df_inplace.fillna({'A':0, 'B':'x', 'C':0}, inplace=True)
print('After inplace fill:')
print(df_inplace)

# Assignment style (preferred for clarity):
df_assigned = df_fill.fillna({'A':0, 'B':'x', 'C':0})
print('After assignment fill:')
print(df_assigned)
```

```python
# 6) Fill categorical using mode (most frequent value)
df_mode = df_fill.copy()
mode_b = df_mode['B'].mode()
mode_val = mode_b.iloc[0] if not mode_b.empty else 'missing'
df_mode['B'] = df_mode['B'].fillna(mode_val)
print('Fill B with mode:')
print(df_mode)

# 7) Use combine_first to fill from another DataFrame (useful when you have a secondary source)
df_primary = pd.DataFrame({'A': [np.nan, 2, np.nan, 4], 'B': [None, 'b', None, None]})
df_secondary = pd.DataFrame({'A': [1, 99, 3, 99], 'B': ['x','y','z','w']})
filled = df_primary.combine_first(df_secondary)
print('combine_first result (primary filled from secondary):')
print(filled)

# 8) Common pattern: fill numeric columns with mean and categorical with mode in one pipeline
df_pipeline = df_fill.copy()
numeric_cols = df_pipeline.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df_pipeline.select_dtypes(exclude=[np.number]).columns.tolist()
df_pipeline[numeric_cols] = df_pipeline[numeric_cols].fillna(df_pipeline[numeric_cols].mean())
for c in categorical_cols:
    m = df_pipeline[c].mode()
    if not m.empty:
        df_pipeline[c] = df_pipeline[c].fillna(m.iloc[0])
print('Pipeline: numeric->mean, categorical->mode')
print(df_pipeline)
```
