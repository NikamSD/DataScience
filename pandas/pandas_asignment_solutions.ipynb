```python
## Assignment 2.1
import numpy as np
import pandas as pd
data2_1 = pd.DataFrame(np.random.randint(0,50, size=(5,3)), columns=['P', 'Q', 'R'])
print("Original DataFrame for Assignment 2.1:")    
print(data2_1)
```

```python
data2_1['S'] = data2_1['P'] * data2_1['Q']
print("\nDataFrame after adding column 'S':")       
print(data2_1)
```

### Alternative solution using fillNa

```python
data2_1['S'] = data2_1[['P','Q']].fillna(data2_1[['P','Q']]).apply(lambda x: x[0] * x[1], axis=1)
print("\nDataFrame after adding column 'S':")
print(data2_1)
```

```python
row_wise_sum = data2_1.sum(axis=1)
print("\nRow-wise sum of the DataFrame:")   
print(row_wise_sum)
```

```python
column_wise_sum = data2_1.sum(axis=0)
print("\nColumn-wise sum of the DataFrame:")    
print(column_wise_sum)
```

```python
## Assignment 6: Time Series Analysis
import pandas as pd 
import numpy as np

date_range = pd.date_range(start='2025-01-01', end='2025-12-31', freq='D')
data6_1 = pd.DataFrame(date_range, columns=['Date'])
print(data6_1)
```

```python
data6_1['Number'] = np.random.randint(1, 100, size=len(data6_1))
data6_1
```

```python
## Resample the data frame to compute the monthly mean
monthly_mean = data6_1.resample('M', on='Date').mean()
monthly_mean
```

```python
date_range_6_2 = pd.date_range(start='2021-01-01', end='2021-12-31', freq='D')
data6_2 = pd.DataFrame(date_range_6_2, columns=['Date'])
data6_2['Value'] = np.random.randint(50, 120, size=len(data6_2))
data6_2.set_index('Date', inplace=True)
data6_2
```

```python
# compute rolling mean with a window of 7 days
rolling_mean_7d = data6_2['Value'].rolling(window=7).mean()
rolling_mean_7d
```
